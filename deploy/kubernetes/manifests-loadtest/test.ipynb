```python
# ============================================================
# CÉLULA 1 — Imports e Configuração
# ============================================================
import requests
import base64
from time import sleep

BASE_URL = "http://100.73.36.16:8080"
TIMEOUT = 10

# Preencha aqui
USERNAME = "teste2"
PASSWORD = "123456"

session = requests.Session()


def show(resp):
    print("STATUS:", resp.status_code)
    print("URL:", resp.url)
    print("COOKIES:", session.cookies.get_dict())
    print("\nHEADERS:")
    print(dict(resp.headers))
    print("\nBODY:")
    print(resp.text[:3000])
```

```python
# ============================================================
# CÉLULA 2 — Login (GET /login com Basic Auth)
# ============================================================
sleep(0.5)

credenciais = f"{USERNAME}:{PASSWORD}"
token = base64.b64encode(credenciais.encode()).decode()

headers = {
    "Authorization": f"Basic {token}",
    "Accept": "*/*",
    "X-Requested-With": "XMLHttpRequest"
}

resp = session.get(
    f"{BASE_URL}/login",
    headers=headers,
    timeout=TIMEOUT
)

show(resp)
print(f"\n✓ Login: {resp.status_code}")
```

```python
# ============================================================
# CÉLULA — Jornada preparatória antes de /address
# ============================================================
sleep(0.5)

rotas = [
    "/index.html",
    "/catalogue",
    "/basket.html",
    "/cart"
]

for rota in rotas:
    sleep(1)
    resp = session.get(f"{BASE_URL}{rota}", timeout=TIMEOUT)
    print(rota, resp.status_code)
```

```python
# ============================================================
# CÉLULA — Testar /address após navegação
# ============================================================
sleep(0.5)

headers = {
    "Accept": "*/*",
    "Referer": f"{BASE_URL}/basket.html",
    "X-Requested-With": "XMLHttpRequest"
}

resp = session.get(
    f"{BASE_URL}/address",
    headers=headers,
    timeout=TIMEOUT
)

show(resp)
```

```python
# ============================================================
# CÉLULA 2 — Ver carrinho antes do checkout (opcional)
# ============================================================
sleep(0.5)

resp = session.get(
    f"{BASE_URL}/cart",
    timeout=TIMEOUT
)

show(resp)
print(f"\n✓ Basket: {resp.status_code}")
```

```python
# ============================================================
# CÉLULA — Listar cartões disponíveis
# ============================================================
sleep(0.5)

resp = session.get(
    f"{BASE_URL}/cards",
    timeout=TIMEOUT
)

show(resp)
print(f"\n✓ Cards: {resp.status_code}")
```

```python
# ============================================================
# CÉLULA 5 — Checkout realista (POST /orders)
# Conforme capturado no navegador: body vazio
# ============================================================
sleep(0.5)

headers = {
    "Accept": "*/*",
    "Origin": BASE_URL,
    "Referer": f"{BASE_URL}/basket.html",
    "X-Requested-With": "XMLHttpRequest"
}

resp = session.post(
    f"{BASE_URL}/orders",
    headers=headers,
    timeout=TIMEOUT
)

show(resp)
print(f"\n✓ Orders POST: {resp.status_code}")
```

```python
# ============================================================
# CÉLULA 6 — Consultar pedidos após compra
# ============================================================
sleep(0.5)

resp = session.get(
    f"{BASE_URL}/orders",
    timeout=TIMEOUT
)

show(resp)
print(f"\n✓ Orders GET: {resp.status_code}")
```
